## 1. Importar librerías

```python
from deepagents import create_deep_agent
```

## 2. Definir Agente

```python
# Importar api key desde las variables de entorno
from dotenv import load_dotenv
load_dotenv()

# Definir un agente con el modelo Gemini 3.5 Flash de Google GenAI
agent = create_deep_agent(
    model="google_genai:gemini-3.5-flash",
    system_prompt="You are a helpful assistant",
)
```

## 3. Probar Agente

```python
query = "¿Qué puedes hacer?"
```

```python
# Run the agent
result = agent.invoke(
        {"messages": [{"role": "user", 
                       "content": query}]}
)
```

```python
# Usar librería rich para imprimir el resultado de manera más legible
from rich import print as rprint
rprint(result)
```

```python
from IPython.display import Markdown, display

# Display as formatted markdown
display(Markdown(result["messages"][-1].content[0]['text']))
```
